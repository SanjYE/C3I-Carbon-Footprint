```python
import matplotlib.pyplot as plt


training_time_hours = [0,0.5,1.0,1.5,2.0,2.5,3.0,3.41]  
energy_emissions = [0,0.179785,0.352383,0.524954,0.696172,0.868292,1.040619,1.183724] 


plt.figure(figsize=(20, 10))
plt.plot(training_time_hours, energy_emissions, marker='.', linestyle='-', color='b')


plt.title('Energy Emissions vs Training Time (DensNet121)')
plt.xlabel('Training Time (Hours)')
plt.ylabel('Energy Emissions')


plt.grid(True)
plt.show()
```
