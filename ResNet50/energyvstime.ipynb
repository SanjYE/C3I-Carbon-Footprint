```python
import matplotlib.pyplot as plt


training_time_hours = [0,0.5,1.0,1.5,2.0,2.5,2.75]  
energy_emissions = [0,0.177354,0.352702,0.533489,0.711723,0.889963,0.981405] 


plt.figure(figsize=(20, 10))
plt.plot(training_time_hours, energy_emissions, marker='.', linestyle='-', color='b')


plt.title('Energy Emissions vs Training Time (ResNet-50)')
plt.xlabel('Training Time (Hours)')
plt.ylabel('Energy Emissions')


plt.grid(True)
plt.show()
```
