```python
import matplotlib.pyplot as plt


training_time_hours = [0,0.25,0.5,0.75,1.0,1.25,1.5,1.75,2.0]  
energy_emissions = [0,0.078794,0.154845,0.231485,0.308302,0.389095,0.472344,0.556292,0.639623] 


plt.figure(figsize=(20, 10))
plt.plot(training_time_hours, energy_emissions, marker='.', linestyle='-', color='b')


plt.title('Energy Emissions vs Training Time (EfficientNet-B0)')
plt.xlabel('Training Time (Hours)')
plt.ylabel('Energy Emissions')


plt.grid(True)
plt.show()
```
