```python
import matplotlib.pyplot as plt


training_time_hours = [0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5]  
energy_emissions = [0,0.184221,0.365888,0.546362,0.725987,0.903706,1.082655,1.260705,1.438675,1.615098] 


plt.figure(figsize=(20, 10))
plt.plot(training_time_hours, energy_emissions, marker='.', linestyle='-', color='b')


plt.title('Energy Emissions vs Training Time (VGG16)')
plt.xlabel('Training Time (Hours)')
plt.ylabel('Energy Emissions')


plt.grid(True)
plt.show()
```
