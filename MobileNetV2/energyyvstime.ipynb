```python
import matplotlib.pyplot as plt


training_time_hours = [0,0.25,0.5,0.75,1.0,1.25,1.5]  
energy_emissions = [0,0.076519,0.153461,0.230838,0.308169,0.385601,0.463020] 


plt.figure(figsize=(20, 10))
plt.plot(training_time_hours, energy_emissions, marker='.', linestyle='-', color='b')


plt.title('Energy Emissions vs Training Time (MobileNetV2)')
plt.xlabel('Training Time (Hours)')
plt.ylabel('Energy Emissions')


plt.grid(True)
plt.show()
```
